```python
def Update(scores, picks):
    cmptBowls = scores[scores['Pick'] != "none"]
    copy = picks[picks['Bowl'].isin(list(cmptBowls['Bowl']))]
    copyMerge = copy.merge(cmptBowls, on='Bowl', suffixes=('_user','_actual'))
    print(copyMerge.head(20))
    
    newScores = []
    for index, row in copyMerge.iterrows():
        if row['Pick_user'] == row['Pick_actual']:
            spreadDiff = np.absolute(row['Spread_user'] - row['Spread_actual'])
            print(f"{row['Name']} - {row['Pick_actual']} {row['Pick_user']} {spreadDiff}")
            if spreadDiff == 0:
                newScores.append(10)
            if spreadDiff <= 3 and spreadDiff != 0:
                newScores.append(8)
            if spreadDiff <= 7 and spreadDiff > 3:
                newScores.append(6)
            if spreadDiff > 7:
                newScores.append(5)
        else:
            newScores.append(0)
    print(f"Length of Picks {len(copyMerge)} --- Length of New Scores {len(newScores)}")
    copyMerge['Score'] = newScores
    print(copyMerge[copyMerge['Name'] == 'Eric Kirchmann'].head(10))
    copyMerge.to_csv('Detail.csv')
    finalScore = copyMerge[['Name', 'Score']]
    finalScore = finalScore.groupby(['Name'], sort=False)['Score'].sum().reset_index()
    sortedFinal = finalScore.sort_values('Score', ascending=False)

    sortedFinal['Rank'] = sortedFinal['Score'].rank(method='min', ascending=False)
    
#     place
#     places = []
#     for index, row in sortedFinal.iterrows():
#         if index == 0:
#             places.append(place)
#         if index != 0:
#             if index 
    stringList = []
    for index, row in sortedFinal.iterrows():
        stringList.append(f"{int(row['Rank'])}. {row['Name']} - {row['Score']}")
    sortedFinal['String'] = stringList
    
    sortedFinal.to_csv('results.csv', index=False)
    print(sortedFinal)
    
        
```

```python
Update(scores, picks)
```
